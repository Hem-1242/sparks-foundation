### **S1 : Import necessary libraries**


```python
import numpy as np
import pandas as pd 
import seaborn as sns
from sklearn import metrics
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

### **S2 : Read data from the source**


> Here, we have student dataset which contains only 2 attributes : **hours and scores.**<br>
Our objective is to predict the percentage of the students based on the number of study hours. <br>
As this task involves just two variables, it is a **Simple Linear Regression** problem.



```python
# Importing data
dataset_url= 'http://bit.ly/w-data'
data = pd.read_csv(dataset_url)
data.head()
```

```python
data.describe()
```

```python
data.info()
```

### **S3 : Data Visualization**

```python
X = np.array(data[['Hours']]) #Input Data
y = np.array(data['Scores'])  #Output Data

# Plotting the score to visualize it on graph 
plt.scatter(X,y)
plt.title('Hours vs Percentage', color='green', size=20)
plt.xlabel('Hours studied',size=15)
plt.ylabel('Percentage scores',size=15)

fig = plt.gcf()
fig.set_size_inches(8,5)
plt.show()
```

##### From the graph above, we can clearly see that there is a positive linear relation between the number of hours studied and percentage of score.

### **S4 : Splitting data for Training and Testing**

```python
#Assigning 80% data for training and 20% for testing
x_train, x_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

### **S5 : Building Model (Linear Regression)**

```python
model = LinearRegression()
model.fit(x_train,y_train)
```

### **S6 : Prediction for Test Data**

```python
y_pred = model.predict(x_test)

#Creating dataframe 
df = pd.DataFrame({'Actual': y_test, 'Predicted': y_pred})
df
```

### **S7 : Evaluation of Model**



```python
print('\n----------------------------  Mean Absolute Error  -----------------------------\n')
print("MAE : ",metrics.mean_absolute_error(y_true=y_test,y_pred=y_pred))

print('\n----------------------------  Mean Squared Error  -----------------------------\n')
print("MSE : ",metrics.mean_squared_error(y_true=y_test,y_pred=y_pred))
```

### **Plotting Regression Line**

```python
line = model.coef_*X+model.intercept_

plt.figure(figsize=(10,5))
plt.scatter(X,y,s=40)
plt.grid(True)
plt.title('Hours vs Percentage', color='green', size=20)
plt.xlabel('Hours studied',size=15)
plt.ylabel('Percentage scores',size=15)
plt.plot(X,line,color='red');
plt.show()
```

### **Sub Task**
> What will be predicted score if a student studies for 9.25 hrs/ day?

```python
hours = 9.25
prediction = model.predict([[hours]])
print("No of Hours : ",hours)
print("Predicted Score : ",prediction[0])
print("\nHence, if a student studies for {0} Hours then he can score {1:.2f}%".format(hours,prediction[0]))
```
